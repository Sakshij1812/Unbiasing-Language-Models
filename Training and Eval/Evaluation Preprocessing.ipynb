##Load dataset

```python
#Loading ir_datasets and all the documents
dataset = ir_datasets.load("msmarco-qna/train")
docs = dataset.docs_store()
```

```python
#Loading eval dataset
with open('/content/drive/MyDrive/Colab Notebooks/Dissertation/Training and Eval/eval_data.pkl', 'rb') as handle:
    data = pickle.load(handle)
data
```

```python
#Creating query-document pair list and target labels list for evaluation 
inp, out = [], []
for index, row in data.iterrows():
  i = 'Query: ' + str(row['query']) + ' Document: ' + str(docs.get(row['docno']).text)
  o = 'true' if row['relevance'] == 1 else 'false'
  inp.append(i)
  out.append(o)
```

```python
#Saving the lists
with open('/content/drive/MyDrive/Colab Notebooks/Dissertation/Training and Eval/inp.pkl', 'wb') as handle:
    pickle.dump(inp, handle)

with open('/content/drive/MyDrive/Colab Notebooks/Dissertation/Training and Eval/out.pkl', 'wb') as handle:
    pickle.dump(out, handle)
```
